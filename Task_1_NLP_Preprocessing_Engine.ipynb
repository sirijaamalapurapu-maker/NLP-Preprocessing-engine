## **1. What is the difference between "Love" and "love" in NLP?**




*   "Love" (capitalized) and "love" (lowercase) are treated as different tokens in raw text.
*   NLP models may consider them as two separate words, which can reduce accuracy.
*   That’s why we usually convert text to lowercase during preprocessing.

### **Task 2: Preprocessing Function**

```python
import re

def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove URLs and emails
    text = re.sub(r'http\S+|www\S+|@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short words (<=2) except 'no', 'not'
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

    # Join tokens
    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

### **Task 3: Stress Testing**

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely loooved this bag 😍😍",
    "Worst food ever... 1/10",
    "Call me at 7458962135",
    "This is THE best internship!!!",
    "Visit https://openai.com now!",
    "Nooooo this is tooooo bad!!!",
    "OK OK I got it",
    "Win 1cr now!!! Limited offer!!!",
    "I am not satisfied with this movie"
]

for text in test_sentences:
    tokens, cleaned = preprocess_text(text)
    print("Original:", text)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned)
    print("-"*50)
```

### **Task 4: Token Analytics**

```python
def token_analysis(tokens):
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(word) for word in tokens) / total if total > 0 else 0
    return total, unique, round(avg_len, 2)

for text in test_sentences:
    tokens, _ = preprocess_text(text)
    total, unique, avg = token_analysis(tokens)

    print("Sentence:", text)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", avg)
    print("-"*50)
```

## **Task 5: Frequency Analysis**

```python
from collections import Counter

all_tokens = []

for text in test_sentences:
    tokens, _ = preprocess_text(text)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

## **Task 6: Full Pipeline**

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

result = full_pipeline(test_sentences)
print(result)
```
